```python
import math 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

t = 100.0 #サンプルする時間の長さ(T)
n = 100 #サンプル数(N)
dt = t/n #サンプリング周期(Δt)

tt = np.arange(n)*dt
f = np.sin(tt) + np.sin(tt*np.sqrt(2.0))
Fk = np.zeros(n) #FFT
ak = np.zeros(n)
bk = np.zeros(n)
omega = np.linspace(0,2*np.pi,100)
Pt = np.zeros(n)
for k in range (n):
    for i in range(n):
       Fk[k] += t*f[i]*np.exp(-(1.j)*2*np.pi*k*i/n)/n
    ak[k] = np.real(Fk[k])
    bk[k] = np.imag(Fk[k])
    Pt[k] = ((ak[k])**2 + (bk[k])**2)/t
plt.plot(omega,Pt)
```
